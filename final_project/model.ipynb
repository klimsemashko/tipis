```
import numpy as np
import pandas as pd
```

```
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import joblib
```

```
df = pd.read_csv('Real_estate.csv', delimiter=',')
```

```
display(df)
```

```
print(df.head())
print(df.info())
print(df.describe())
```

```

df = df[['X1 transaction date', 'X2 house age', 'X3 distance to the nearest MRT station',
         'X4 number of convenience stores', 'X5 latitude', 'X6 longitude', 'Y house price of unit area']]

df = df.dropna()

X = df[['X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores',
        'X5 latitude', 'X6 longitude']]
y = df['Y house price of unit area']
```

```
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MAE: {mae}')
print(f'R²: {r2}')

# Сохраняем модель и скейлер
joblib.dump(model, 'house_price_model.pkl')
joblib.dump(scaler, 'scaler.pkl')

print("Модель и скейлер успешно сохранены.")
```
